```python
"""
Climate NER Pre-Annotation Pipeline
====================================
Automatically pre-annotates climate abstracts using curated term lists
for 8 entity types. Produces a JSON file ready for import into Label Studio.

Usage:
    python preannotate.py --input climate_abstracts.csv --output preannotated.json

Requirements:
    pip install spacy pandas
    python -m spacy download en_core_web_sm
"""

import json
import re
import argparse
import pandas as pd
import spacy
from spacy.matcher import PhraseMatcher

# ─────────────────────────────────────────────────────────────────────────────
# TERM LISTS  (8 entity types)
# Sources: IPCC AR6 Glossary (Annex VII), IPCC SR1.5 Glossary,
#          ENVO Environment Ontology, ChEBI Chemical Ontology
# ─────────────────────────────────────────────────────────────────────────────

ENTITY_TERMS = {

    # ── 1. CLIMATE DRIVER ────────────────────────────────────────────────────
    # Greenhouse gases, aerosols, and other radiative forcing agents
    "CLIMATE_DRIVER": [
        # Greenhouse gases (ChEBI-derived)
        "carbon dioxide", "CO2", "CO₂",
        "methane", "CH4", "CH₄",
        "nitrous oxide", "N2O", "N₂O",
        "ozone", "O3", "O₃",
        "water vapour", "water vapor",
        "sulfur dioxide", "SO2", "SO₂",
        "nitrogen dioxide", "NO2", "NO₂",
        "hydrofluorocarbons", "HFCs",
        "perfluorocarbons", "PFCs",
        "sulfur hexafluoride", "SF6",
        "chlorofluorocarbons", "CFCs",
        "tropospheric ozone",
        "black carbon", "soot",
        "aerosols", "particulate matter",
        # Forcing concepts (IPCC AR6 Glossary)
        "greenhouse gas", "greenhouse gases", "GHG", "GHGs",
        "radiative forcing", "effective radiative forcing", "ERF",
        "anthropogenic forcing", "climate forcing",
        "carbon emissions", "CO2 emissions", "CO₂ emissions",
        "greenhouse gas emissions", "GHG emissions",
        "fossil fuel emissions", "carbon pollution",
        "methane emissions", "methane flux",
        "permafrost methane", "wetland methane",
        "atmospheric CO2", "atmospheric carbon dioxide",
        "atmospheric methane", "atmospheric concentration",
        "carbon concentration", "greenhouse effect",
        "short-lived climate forcers", "short-lived climate pollutants",
        "albedo", "surface albedo", "ice-albedo feedback",
        "solar radiation", "solar irradiance", "insolation",
        "volcanic aerosols", "stratospheric aerosols",
    ],

    # ── 2. CLIMATE VARIABLE ───────────────────────────────────────────────────
    # Physical quantities that characterise the climate system (IPCC AR6 WG1)
    "CLIMATE_VARIABLE": [
        # Temperature
        "global temperature", "global mean temperature",
        "global surface temperature", "global mean surface temperature",
        "surface temperature", "air temperature", "land temperature",
        "ocean temperature", "sea surface temperature", "SST",
        "warming", "global warming", "temperature increase",
        "temperature rise", "temperature anomaly", "heat content",
        "ocean heat content", "upper ocean heat content",
        "warming trend", "temperature trend",
        # Precipitation & water cycle
        "precipitation", "rainfall", "snowfall", "runoff",
        "evapotranspiration", "evaporation", "soil moisture",
        "streamflow", "river discharge", "groundwater",
        "freshwater availability", "water cycle",
        "hydrological cycle", "monsoon", "monsoon rainfall",
        # Sea level
        "sea level", "sea level rise", "sea level change",
        "global mean sea level", "relative sea level",
        "sea level increase", "coastal inundation",
        # Ocean variables
        "ocean pH", "ocean acidification", "ocean salinity",
        "ocean circulation", "thermohaline circulation",
        "AMOC", "Atlantic overturning circulation",
        "ocean oxygen", "ocean deoxygenation",
        # Cryosphere
        "sea ice extent", "sea ice area", "sea ice thickness",
        "Arctic sea ice", "Antarctic sea ice",
        "snow cover", "snow depth", "permafrost temperature",
        "permafrost thaw", "active layer thickness",
        # Atmospheric
        "atmospheric pressure", "humidity", "relative humidity",
        "wind speed", "wind pattern", "jet stream",
        "atmospheric circulation", "stratospheric temperature",
        "tropopause", "heat wave frequency", "cold spell",
        # Carbon cycle
        "carbon flux", "carbon uptake", "carbon sink",
        "carbon source", "net ecosystem production",
        "gross primary production", "GPP",
        "terrestrial carbon sink", "ocean carbon sink",
        # Extremes
        "extreme heat", "extreme precipitation", "heavy rainfall",
        "flood frequency", "drought frequency", "fire weather",
    ],

    # ── 3. ENVIRONMENTAL EVENT ────────────────────────────────────────────────
    # Discrete climate-related events and hazards (IPCC AR6 WG1 Ch.11, WG2)
    "ENVIRONMENTAL_EVENT": [
        # Heat events
        "heat wave", "heatwave", "heat stress", "heat extreme",
        "extreme heat event", "hot spell", "urban heat island",
        # Cold events
        "cold wave", "cold snap", "frost", "freeze",
        # Drought & fire
        "drought", "flash drought", "agricultural drought",
        "hydrological drought", "meteorological drought",
        "wildfire", "forest fire", "bushfire", "wildland fire",
        "fire weather", "fire season",
        # Flooding
        "flooding", "flood", "flash flood", "coastal flooding",
        "riverine flooding", "compound flooding", "inundation",
        "storm surge", "tidal flooding", "nuisance flooding",
        # Storms
        "hurricane", "typhoon", "cyclone", "tropical cyclone",
        "tropical storm", "extratropical cyclone",
        "tornado", "windstorm", "storm", "extreme storm",
        "bomb cyclone",
        # Cryosphere events
        "glacier melt", "glacial retreat", "ice sheet collapse",
        "ice shelf collapse", "permafrost thaw", "permafrost collapse",
        "thermokarst", "glacial lake outburst flood", "GLOF",
        "Arctic sea ice loss", "Arctic sea ice decline",
        "Antarctic ice loss",
        # Ocean events
        "coral bleaching", "mass bleaching event",
        "marine heatwave", "harmful algal bloom", "algal bloom",
        "dead zone", "hypoxic event", "ocean dead zone",
        "red tide",
        # Land degradation
        "landslide", "mudslide", "erosion", "desertification",
        "land degradation", "soil erosion", "dust storm",
        "sandstorm",
    ],

    # ── 4. ECOSYSTEM ──────────────────────────────────────────────────────────
    # Biomes, habitats, and ecological systems (ENVO-derived)
    "ECOSYSTEM": [
        # Terrestrial forest
        "tropical forest", "tropical rainforest", "rainforest",
        "Amazon rainforest", "Amazon basin", "Amazon",
        "boreal forest", "taiga", "temperate forest",
        "deciduous forest", "coniferous forest",
        "cloud forest", "montane forest",
        # Other terrestrial
        "grassland", "savanna", "shrubland", "tundra",
        "arctic tundra", "alpine tundra",
        "wetland", "peatland", "bog", "marsh", "swamp",
        "mangrove", "mangrove forest", "mangroves",
        "desert", "arid land", "semi-arid",
        "dryland", "Mediterranean ecosystem",
        # Freshwater
        "freshwater ecosystem", "river ecosystem", "lake ecosystem",
        "stream", "river delta", "floodplain", "estuary",
        # Cryosphere ecosystems
        "glacier", "glaciers", "ice sheet", "ice sheets",
        "Greenland ice sheet", "Antarctic ice sheet",
        "sea ice ecosystem", "permafrost ecosystem",
        "Arctic ecosystem", "Antarctic ecosystem",
        "cryosphere",
        # Ocean & coastal
        "coral reef", "coral reefs", "reef ecosystem",
        "Great Barrier Reef",
        "kelp forest", "seagrass meadow", "seagrass",
        "salt marsh", "coastal wetland", "intertidal zone",
        "deep sea", "deep ocean", "open ocean",
        "coastal ecosystem", "marine ecosystem",
        "pelagic zone", "benthic zone",
        # Mountains
        "mountain ecosystem", "alpine ecosystem", "highland",
        "Himalayan ecosystem", "Andean ecosystem",
    ],

    # ── 5. SPECIES ────────────────────────────────────────────────────────────
    # Organisms and taxonomic groups mentioned in climate impact literature
    "SPECIES": [
        # Marine
        "coral", "corals", "coral species",
        "fish", "marine fish", "reef fish",
        "plankton", "phytoplankton", "zooplankton",
        "krill", "marine mammals", "whale", "dolphin", "seal",
        "sea turtle", "marine turtle",
        "polar bear", "walrus", "narwhal",
        "seabird", "penguin", "albatross",
        "oyster", "shellfish", "bivalve", "mollusc",
        "sea urchin", "starfish", "echinoderm",
        # Terrestrial mammals
        "polar bear", "Arctic fox", "lemming",
        "reindeer", "caribou", "moose",
        "elephant", "lion", "cheetah",
        "migratory bird", "songbird", "raptor",
        # Plants & trees
        "tree species", "boreal tree", "spruce", "pine",
        "deciduous tree", "mangrove species",
        "alpine plant", "tundra plant",
        "crop", "crops", "wheat", "maize", "rice", "soybean",
        # Insects & invertebrates
        "insect", "beetle", "butterfly", "bee",
        "pollinator", "monarch butterfly",
        # Amphibians & reptiles
        "amphibian", "frog", "salamander",
        "reptile", "lizard",
        # Fungi & microbes
        "fungi", "soil microbe", "microbiome",
        # General
        "species", "endemic species", "native species",
        "invasive species", "migratory species",
        "keystone species", "apex predator",
    ],

    # ── 6. HUMAN ACTIVITY ─────────────────────────────────────────────────────
    # Anthropogenic drivers and land-use activities (IPCC AR6 WG3)
    "HUMAN_ACTIVITY": [
        # Fossil fuels
        "fossil fuel burning", "fossil fuel combustion",
        "coal burning", "coal combustion", "coal mining",
        "oil extraction", "oil production", "fracking",
        "natural gas extraction", "gas flaring",
        "fossil fuel use", "fossil fuel consumption",
        # Land use
        "deforestation", "forest clearing", "forest loss",
        "logging", "illegal logging",
        "land use change", "land cover change",
        "land degradation", "overgrazing",
        "urbanisation", "urbanization", "urban expansion",
        "agricultural expansion", "cropland expansion",
        # Agriculture
        "agriculture", "farming", "livestock farming",
        "rice cultivation", "paddy cultivation",
        "fertiliser use", "nitrogen fertiliser",
        "irrigation", "intensive farming",
        # Industry & transport
        "industrialisation", "industrialization",
        "industrial emissions", "manufacturing",
        "transportation emissions", "aviation emissions",
        "shipping emissions", "road transport",
        # Waste & pollution
        "waste management", "landfill",
        "plastic pollution", "ocean pollution",
        "air pollution", "water pollution",
        # Mitigation activities
        "reforestation", "afforestation", "tree planting",
        "forest restoration", "ecosystem restoration",
        "carbon capture", "carbon sequestration",
        "direct air capture", "DAC",
        "bioenergy", "BECCS",
        "solar energy", "solar power", "wind energy", "wind power",
        "renewable energy", "clean energy",
        "energy efficiency", "electrification",
        "sustainable agriculture", "conservation agriculture",
    ],

    # ── 7. GEOGRAPHIC LOCATION ────────────────────────────────────────────────
    # Regions, places, and geographic features relevant to climate (IPCC AR6)
    "GEOGRAPHIC_LOCATION": [
        # Polar regions
        "Arctic", "Antarctic", "Antarctica",
        "North Pole", "South Pole",
        "Greenland", "Iceland",
        "Siberia", "Alaska",
        # Mountain ranges
        "Himalayas", "Hindu Kush", "Karakoram",
        "Alps", "Andes", "Rocky Mountains",
        "Tibetan Plateau", "Hindu Kush Himalaya",
        # Oceans & seas
        "Pacific Ocean", "Atlantic Ocean", "Indian Ocean",
        "Arctic Ocean", "Southern Ocean",
        "Mediterranean Sea", "Caribbean Sea",
        "Coral Triangle", "Gulf of Mexico",
        "Bay of Bengal", "Arabian Sea",
        # Tropical regions
        "Amazon", "Congo Basin", "Mekong",
        "Southeast Asia", "South Asia", "East Asia",
        "Sub-Saharan Africa", "Sahel", "Horn of Africa",
        "Small Island Developing States", "SIDS",
        "Pacific Islands", "Caribbean",
        # Coastal regions
        "Bangladesh", "Maldives", "Tuvalu",
        "Netherlands", "Florida", "Louisiana",
        "Ganges delta", "Mekong delta", "Nile delta",
        "coastal regions", "low-lying coastal areas",
        # General regions
        "tropics", "tropical regions", "subtropics",
        "boreal region", "temperate region",
        "arid region", "semi-arid region",
        "global", "worldwide", "global scale",
        "Northern Hemisphere", "Southern Hemisphere",
    ],

    # ── 8. POLICY ─────────────────────────────────────────────────────────────
    # Climate agreements, regulations, and mitigation frameworks (IPCC AR6 WG3)
    "POLICY": [
        # International agreements
        "Paris Agreement", "Paris accord",
        "Kyoto Protocol", "UNFCCC",
        "United Nations Framework Convention on Climate Change",
        "COP26", "COP27", "COP28", "COP21",
        "Glasgow Climate Pact",
        "Montreal Protocol",
        # Targets & commitments
        "net zero", "net-zero emissions", "carbon neutrality",
        "carbon neutral", "climate neutrality",
        "1.5 degree target", "1.5°C target", "2 degree target", "2°C target",
        "emission reduction target", "NDC",
        "nationally determined contribution",
        "nationally determined contributions",
        "long-term strategy", "carbon budget",
        # Carbon pricing
        "carbon tax", "carbon pricing", "carbon price",
        "emissions trading", "cap and trade", "cap-and-trade",
        "carbon market", "carbon credit", "carbon offset",
        "EU ETS", "emissions trading scheme",
        # Finance & investment
        "climate finance", "green finance",
        "climate investment", "green bond",
        "loss and damage", "adaptation finance",
        "climate fund", "Green Climate Fund",
        # Policy instruments
        "emission regulation", "emissions regulation",
        "renewable energy policy", "feed-in tariff",
        "carbon standard", "climate legislation",
        "climate policy", "climate action",
        "mitigation policy", "adaptation policy",
        "just transition", "climate justice",
        "nature-based solutions", "NbS",
        "ecosystem-based adaptation",
    ],
}


# ─────────────────────────────────────────────────────────────────────────────
# PIPELINE
# ─────────────────────────────────────────────────────────────────────────────

def build_matcher(nlp):
    matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
    for label, terms in ENTITY_TERMS.items():
        patterns = [nlp.make_doc(t.lower()) for t in terms]
        matcher.add(label, patterns)
    return matcher


def annotate_text(text, nlp, matcher):
    """Return list of {start, end, label} spans for one text."""
    doc = nlp(text)
    matches = matcher(doc)

    # Convert to character offsets; resolve overlaps (keep longer span)
    spans = []
    for match_id, start, end in matches:
        label = nlp.vocab.strings[match_id]
        span = doc[start:end]
        spans.append({
            "start": span.start_char,
            "end": span.end_char,
            "label": label,
            "text": span.text
        })

    # Remove overlapping spans – keep the longest one
    spans = sorted(spans, key=lambda s: s["end"] - s["start"], reverse=True)
    kept = []
    occupied = set()
    for span in spans:
        chars = set(range(span["start"], span["end"]))
        if not chars & occupied:
            kept.append(span)
            occupied |= chars

    return sorted(kept, key=lambda s: s["start"])


def to_label_studio_task(row, spans):
    """Convert one abstract + its spans to a Label Studio JSON task."""
    results = []
    for i, span in enumerate(spans):
        results.append({
            "id": f"entity_{i}",
            "type": "labels",
            "value": {
                "start": span["start"],
                "end": span["end"],
                "text": span["text"],
                "labels": [span["label"]]
            },
            "from_name": "label",
            "to_name": "text",
            "origin": "prediction"
        })

    return {
        "data": {
            "text": row["abstract"],
            "title": row.get("title", ""),
            "year": str(row.get("year", "")),
        },
        "predictions": [{
            "model_version": "dict_preannotation_v1",
            "result": results,
            "score": 0.0
        }]
    }


def run(input_csv, output_json, max_rows=None):
    print(f"Loading spaCy model...")
    nlp = spacy.load("en_core_web_sm", disable=["ner", "parser", "tagger"])

    print(f"Building PhraseMatcher with {sum(len(v) for v in ENTITY_TERMS.values())} terms across 8 entity types...")
    matcher = build_matcher(nlp)

    print(f"Reading {input_csv}...")
    df = pd.read_csv(input_csv)
    if max_rows:
        df = df.head(max_rows)

    # Print term counts per entity type
    print("\nTerm counts per entity type:")
    for label, terms in ENTITY_TERMS.items():
        print(f"  {label:25s}: {len(terms)} terms")

    tasks = []
    total_spans = 0
    empty = 0

    print(f"\nAnnotating {len(df)} abstracts...")
    for i, row in df.iterrows():
        abstract = str(row.get("abstract", ""))
        if not abstract.strip():
            empty += 1
            continue
        spans = annotate_text(abstract, nlp, matcher)
        total_spans += len(spans)
        tasks.append(to_label_studio_task(row, spans))

        if (i + 1) % 100 == 0:
            print(f"  Processed {i+1}/{len(df)} abstracts...")

    print(f"\nDone.")
    print(f"  Abstracts annotated : {len(tasks)}")
    print(f"  Empty abstracts     : {empty}")
    print(f"  Total entity spans  : {total_spans}")
    print(f"  Avg spans/abstract  : {total_spans/max(len(tasks),1):.1f}")

    with open(output_json, "w", encoding="utf-8") as f:
        json.dump(tasks, f, indent=2, ensure_ascii=False)
    print(f"\nSaved Label Studio import file: {output_json}")

    # Print entity type distribution
    from collections import Counter
    label_counts = Counter()
    for task in tasks:
        for r in task["predictions"][0]["result"]:
            label_counts[r["value"]["labels"][0]] += 1

    print("\nEntity type distribution in corpus:")
    for label, count in sorted(label_counts.items(), key=lambda x: -x[1]):
        print(f"  {label:25s}: {count}")

    return tasks


# ─────────────────────────────────────────────────────────────────────────────
# ALSO EXPORT TERM LISTS AS CSV (for transparency / appendix)
# ─────────────────────────────────────────────────────────────────────────────

def export_term_lists(output_csv):
    rows = []
    for label, terms in ENTITY_TERMS.items():
        for term in terms:
            rows.append({"entity_type": label, "term": term})
    df = pd.DataFrame(rows)
    df.to_csv(output_csv, index=False)
    print(f"Term lists exported: {output_csv} ({len(rows)} terms total)")
```

```python
# Replace the if __name__ == "__main__": block with this

import subprocess
subprocess.run(["python", "-m", "spacy", "download", "en_core_web_sm"])

# Set your paths directly here
INPUT_CSV  = r"climate_abstracts.csv"
OUTPUT_JSON = r"D:\Climate_Intelligent_System\data\annotated\preannotated.json"
TERMS_CSV  = r"term_lists.csv"

export_term_lists(TERMS_CSV)
run(INPUT_CSV, OUTPUT_JSON)
```
